## 7.Crear el gráfico de torta
* Investigar en la página de [Matplotlib](https://matplotlib.org/) cuál es el parámetro que cambia para hacer un gráfico de torta
* ¿Qué otros tipos de gráficos se pueden realizar?

respuesta: Gridded data,Statistical distributions,Pairwise data and Irregularly gridded data

````python
#completar
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
datos_aeropuertos_argentina = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSyDciqZ_jkvAtX-11_p-hup0lYHoOmlpiQsA8cqZJEEWDMz4cOeytExpxESHk09WCaAcZnMvHt76ED/pub?gid=294700430&single=true&output=csv')
conteo_tipos = datos_aeropuertos_argentina['type'].value_counts()
tipos = conteo_tipos.index
valores = conteo_tipos.values
colores = plt.cm.Blues(np.linspace(0.2, 0.7, len(tipos)))
plt.figure(figsize=(10, 6))
wedges, texts, autotexts = plt.pie(valores, labels=None, colors=colores, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.pie(valores, labels=None, colors=colores, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.setp(autotexts, size=10, weight="bold")
plt.setp(texts, size=10)
plt.title('Distribución de Tipos de Aeropuertos en Argentina')
plt.legend(tipos, loc='best', bbox_to_anchor=(1, 0.5))
plt.show()


````
